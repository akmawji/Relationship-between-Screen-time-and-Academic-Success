```python
# General Imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# CSV File Import

from google.colab import files
uploaded = files.upload()

# Import Dataset
csv_file_name = list(uploaded.keys())[0]
df = pd.read_csv(csv_file_name)

# Refactoring Columns
newColumns = ['Timestamp',
              'Are you an engineering student?',
              'Average Daily Screen Time',
              'Most Active Time on Phone',
              'GPA (4.3 Scale)']

df.columns = newColumns

# Display Dataset
display(df)

# Distribution of Average Daily Screen Time
plt.figure(figsize=(10, 6))
sns.countplot(y=df['Average Daily Screen Time'])
plt.title('Distribution of Average Daily Screen Time')
plt.xlabel('Count')
plt.ylabel('Average Daily Screen Time')
plt.show()

# Distribution of Most Active Time on Phone
plt.figure(figsize=(10, 6))
sns.countplot(y=df['Most Active Time on Phone'])
plt.title('Distribution of Most Active Time on Phone')
plt.xlabel('Count')
plt.ylabel('Most Active Time on Phone')
plt.show()

# Relationship between Average Daily Screen Time and GPA
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Average Daily Screen Time'], y=df['GPA (4.3 Scale)'])
plt.title('Relationship between Average Daily Screen Time and GPA')
plt.xlabel('Average Daily Screen Time')
plt.ylabel('GPA (4.3 Scale)')
plt.show()

# Relationship between Most Active Time on Phone and GPA
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Most Active Time on Phone'], y=df['GPA (4.3 Scale)'])
plt.title('Relationship between Most Active Time on Phone and GPA')
plt.xlabel('Most Active Time on Phone')
plt.ylabel('GPA (4.3 Scale)')
plt.xticks(rotation=45)
plt.show()

# General Analysis for GPA (4.3 Scale)
mean_gpa = df['GPA (4.3 Scale)'].mean()
std_gpa = df['GPA (4.3 Scale)'].std()
var_gpa = df['GPA (4.3 Scale)'].var()
min_gpa = df['GPA (4.3 Scale)'].min()
max_gpa = df['GPA (4.3 Scale)'].max()
median_gpa = df['GPA (4.3 Scale)'].median()

print('General Analysis for GPA (4.3 Scale):')
print(f'Mean: {mean_gpa}')
print(f'Standard Deviation: {std_gpa}')
print(f'Variance: {var_gpa}')
print(f'Min: {min_gpa}')
print(f'Max: {max_gpa}')
print(f'Median: {median_gpa}')
print('\n')

# Confidence Interval for the Population Mean
confidence_level = 0.95
degrees_freedom = len(df) - 1
sample_mean = mean_gpa
sample_standard_error = std_gpa / np.sqrt(len(df))
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

print(f'95% Confidence Interval for the Population Mean: {confidence_interval}')

# Approximate Population Variance
population_variance = var_gpa * (len(df) / (len(df) - 1))  # Unbiased estimator

print(f'Approximate Population Mean: {mean_gpa}')
print(f'Approximate Population Variance: {population_variance}')

# T-test: Compare GPA for less than 6 hours vs 6 or more hours of screen time
# Splitting the data into two groups
less_than_6_hours = df[df['Average Daily Screen Time'].isin(['< 1 hour', '1-3 hours', '3-5 hours'])]
more_than_6_hours = df[df['Average Daily Screen Time'].isin(['6-8 hours', '9-11 hours', '12+ hours'])]

# Conducting the t-test
t_stat, p_value = stats.ttest_ind(less_than_6_hours['GPA (4.3 Scale)'], more_than_6_hours['GPA (4.3 Scale)'], nan_policy='omit')

print('\nT-test Results:')
print(f't-statistic: {t_stat}')
print(f'p-value: {p_value}')

```
